```python
exmaples = ["This movie is not good at all!", 
            "I have never seen such a bad film.", 
            "The plot was terrible and I don't recommend it.",
            "I can't say this movie is worth watching."
            "That's great!!!"]
```

```python
import string

def remove_punctuation(text,keep_important=True):
    if keep_important:
        #Keep '!' and '?' as they can carry sentiment information
        punctuation_to_remove = string.punctuation.replace('!', '').replace('?', '')
    else:
        punctuation_to_remove = string.punctuation
    return text.translate(str.maketrans('', '', punctuation_to_remove))
```

```python
sentence = "Wow!!! This movie is amazing, I can't believe it..."

print("Original sentence:")
print(sentence)

print("\nSentence after removing all punctuation:")
print(remove_punctuation(sentence, keep_important=False))

print("\nSentence after removing punctuation (keeping '!' and '?'):")
print(remove_punctuation(sentence, keep_important=True))
```

Stemming & Lemmatization

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

words = ["running", "ran", "runs", "easily", "fairly", "better", "best", "happier", "happiest", "studies", "studying", "studied","wolves","feet"]

for word in words:
    print(f"Original: {word:10} | Stemmed: {stemmer.stem(word):10} | Lemmatized: {lemmatizer.lemmatize(word):10}")  
```

```python
def preprocess_text(text,
                    lowercase=True,
                    normalize=True,
                    remove_stopwords=True,
                    perserve_negation=True,
                    stem=False,
                    lemmatize=False):
    
    if normalize:
        text = re.sub(r'(https?://\S+|www\.\S+)', ' ', text)
        text = re.sub(r'\S+@\S+\.\S+', ' ', text)
        text = re.sub(r'<[^>]+>', ' ', text) # Remove HTML tags

    if lowercase:
        text = text.lower()

    tokens = word_tokenize(text)

    if remove_stopwords:
        stop_words = set(stopwords.words('english'))
        if perserve_negation:
            negation_words = ["not", "no", "never", "n't", "cannot", "can't", "don't", "doesn't", "didn't", "won't", "wouldn't", "shouldn't", "couldn't", "isn't", "aren't", "wasn't", "weren't", "haven't", "hasn't", "hadn't"]
            stop_words = stop_words - set(negation_words)
        tokens = [word for word in tokens if word not in stop_words]    
    
    tokens = [t for t in tokens if t.isalpha() or t == "n't"]
        
    if stem:
        stemmer = PorterStemmer()
        tokens = [stemmer.stem(word) for word in tokens]

    if lemmatize:
        lemmatizer = WordNetLemmatizer()
        tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return tokens
```

```python
sample_text = df['review'].iloc[0]
print("Original Text:") 
print(sample_text)

token_basic = preprocess_text(sample_text)
print("\nPreprocessed Text (Basic):")
print(token_basic[:20])  # Print first 20 tokens after basic preprocessing

token_stemmed = preprocess_text(sample_text, lowercase=True, remove_stopwords=True, perserve_negation=True, stem=True, lemmatize=False)
print("\nPreprocessed Text with Stemming:")
print(token_stemmed[:20])  # Print first 20 tokens after stemming

token_lemmatized = preprocess_text(sample_text, lowercase=True, remove_stopwords=True, perserve_negation=True, stem=False, lemmatize=True)
print("\nPreprocessed Text with Lemmatization:")
print(token_lemmatized)

"""token = word_tokenize(preprocessed_text)
print("Tokens after preprocessing:")
print(token)"""
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer()
```

```python
def clean_text(text):
    
    text = re.sub(r'<[^>]+>', '', text)  # Remove HTML tags
    text = text.lower()  # Convert to lowercase
    return text
```

```python
df['clean_text'] = df['review'].apply(clean_text)
```

```python
vectorizer = CountVectorizer(max_features=1000, stop_words='english')

X_bow = vectorizer.fit_transform(df['clean_text'])
```

```python
print("Shape of BoW feature matrix:", X_bow.shape)
print("Sample feature names:", vectorizer.get_feature_names_out()[:20])
print("Sample BoW vector for first review:")
print(X_bow[0].toarray())
```

```python
print(f"{X_bow.shape[0]} documents and {X_bow.shape[1]} features.")
```

```python
word_counts = X_bow.sum(axis=0).A1
word_freq = list(zip(vectorizer.get_feature_names_out(), word_counts))
word_freq.sort(key=lambda x: x[1], reverse=True)

print("Top 20 most common words:") 
for word, count in word_freq[:20]:
    print(f"{word}: {count}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(
    max_features=1000, 
    stop_words='english',
    ngram_range=(1,2))

X_tidf = tfidf_vectorizer.fit_transform(df['clean_text'])
```

```python
print(f"TF-IDF feature matrix shape: {X_tidf.shape}")
print(f" {X_tidf.shape[0]} documents and {X_tidf.shape[1]} features.")
```

```python
idf_scores = tfidf_vectorizer.idf_
feature_names = tfidf_vectorizer.get_feature_names_out()
```

```python
idf_df = pd.DataFrame({'feature': feature_names, 'idf_score': idf_scores})
idf_df = idf_df.sort_values(by='idf_score', ascending=False)

print("Top 20 features with highest IDF scores:")
print(idf_df.head(20))

print("Top 20 features with lowest IDF scores:")
print(idf_df.tail(20))
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X_tidf, df['sentiment'], test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)

model.fit(x_train, y_train)

model.score(x_test, y_test)
```
